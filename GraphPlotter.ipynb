# Graph Plotter
#### This script plots simply graphs given y and x values

### Imports

```python
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
```

### Define the function to plot

```python
def plot(x, y, x_label, y_label, tick_amount):
    fig, ax = plt.subplots(figsize=(7,5))
    plt.plot(x, y)
    plt.scatter(x, y)
    ax.axes.set_xlabel(x_label)
    ax.axes.set_ylabel(y_label)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_amount))
    ax.set_xlim(left=1, right=max(x))
    ax.set_ylim(bottom=0)
    plt.show(ax)
```
